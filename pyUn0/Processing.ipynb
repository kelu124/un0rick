```python
import glob
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt

import pyUn0 as US

from scipy.signal import hilbert
```

```python
for jsonfile in glob.glob("data/*.json"):
    print(jsonfile)
```

#Â Processing and filtering the signals

```python
y = US.us_json()
y.JSONprocessing(jsonfile)
y.fPiezo = 3.2    # Central frequency of the piezo
y.Bandwidth = 0.7 # Bandwidth of interest
y.create_fft() 
```

```python
y.mkImg()
```

```python
y.EnvHil = np.asarray(np.abs(hilbert(y.filtered_signal)))

plt.figure(figsize=(15, 5))
plt.plot(y.t, y.tmp, 'b-',label="Raw",alpha=0.8)
plt.plot(y.t, y.filtered_signal, 'y-',label="Filtered",alpha=0.8)
plt.plot(y.t,y.EnvHil,label="Enveloppe")
plt.title("Comparing")
plt.xlabel('Freq (MHz)')
plt.tight_layout()
plt.legend()
plt.show()
```
